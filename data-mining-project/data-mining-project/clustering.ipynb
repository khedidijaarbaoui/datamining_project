**Identify Seasonal Clusters with Similar Sales Patterns Across FamilleArticle**  

Our objective is to analyze sales trends for each product family across different seasons to determine the optimal period for maximizing sales. By leveraging clustering as a data mining technique, we aim to group seasons where specific FamilleArticle exhibit similar sales characteristics. This approach allows us to uncover patterns and identify the season that yields the highest sales for each family. Conducting clustering analysis will help us assess the average sales within each cluster across various seasons.

```python
# Select features for clustering
selected_columns = ['FamilleArticle', 'Season']
selected_data = clustering_data[selected_columns]
row_counts = selected_data.groupby(['FamilleArticle', 'Season']).size().reset_index(name='RowCounts')
pivot_data = row_counts.pivot_table(index='FamilleArticle', columns='Season', values='RowCounts', aggfunc='mean', fill_value=0)

# Normalize the data
scaler = StandardScaler()
normalized_data = scaler.fit_transform(pivot_data)

# Function to determine the optimal k using Elbow Method and Silhouette Score
def plot_k_selection(data, max_k=10):
    distortions = []
    silhouette_scores = []
    K_range = range(2, max_k + 1)

    for k in K_range:
        kmeans = KMeans(n_clusters=k, random_state=42)
        labels = kmeans.fit_predict(data)
        distortions.append(kmeans.inertia_)
        silhouette_scores.append(silhouette_score(data, labels))

    # Plot Elbow Method
    plt.figure(figsize=(10, 5))
    
    plt.subplot(1, 2, 1)
    plt.plot(K_range, distortions, 'bo-', markersize=8)
    plt.xlabel('Number of Clusters (k)')
    plt.ylabel('Distortion (Inertia)')
    plt.title('Elbow Method for Optimal k')

    # Plot Silhouette Scores
    plt.subplot(1, 2, 2)
    plt.plot(K_range, silhouette_scores, 'ro-', markersize=8)
    plt.xlabel('Number of Clusters (k)')
    plt.ylabel('Silhouette Score')
    plt.title('Silhouette Score for Optimal k')
    
    plt.show()
plot_k_selection(normalized_data)
```

```python
chosen_k = 4 
```

```python
# K-Means Clustering
kmeans = KMeans(n_clusters=chosen_k, random_state=42)
kmeans_labels = kmeans.fit_predict(normalized_data)
pivot_data['KMeans_Cluster'] = kmeans_labels
kmeans_silhouette = silhouette_score(normalized_data, kmeans_labels)
```

```python
# Hierarchical Clustering
hierarchical_model = AgglomerativeClustering(n_clusters=chosen_k, metric='euclidean', linkage='ward')
hierarchical_labels = hierarchical_model.fit_predict(normalized_data)
pivot_data['Hierarchical_Cluster'] = hierarchical_labels
hierarchical_silhouette = silhouette_score(normalized_data, hierarchical_labels)
# Dendrogram for Hierarchical Clustering
plt.figure(figsize=(10, 5))
z = linkage(normalized_data, method='ward')
dendrogram(z, labels=pivot_data.index, leaf_rotation=90, leaf_font_size=10)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('FamilleArticle')
plt.ylabel('Distance')
plt.show()
```

```python
# Apply DBSCAN Clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)  
dbscan_labels = dbscan.fit_predict(normalized_data)
pivot_data['DBSCAN_Cluster'] = dbscan_labels

# Check if DBSCAN produced valid clusters before computing silhouette score
if len(set(dbscan_labels)) > 1: 
    dbscan_silhouette = silhouette_score(normalized_data, dbscan_labels)
else:
    dbscan_silhouette = "Not applicable (only one cluster detected)"
```

```python
# Display results
print(f"KMeans Silhouette Score: {kmeans_silhouette}")
print(f"Hierarchical Clustering Silhouette Score: {hierarchical_silhouette}")
print(f"DBSCAN Silhouette Score: {dbscan_silhouette}")

print("\nKMeans Clusters:\n", pivot_data['KMeans_Cluster'].value_counts())
print("\nHierarchical Clusters:\n", pivot_data['Hierarchical_Cluster'].value_counts())
print("\nDBSCAN Clusters:\n", pivot_data['DBSCAN_Cluster'].value_counts())
```

This information provides insights into the clustering performance and distribution of different clusters based on the chosen algorithms.  

### **Silhouette Score Analysis**  
The **Kmeans algorithm achieved the highest silhouette score (0.78450)**, indicating the best-defined clusters with the highest cohesion and separation. **hierarchical followed closely (0.75829)**, while **DBSACN Clustering had the lowest score (0.5577)**, suggesting slightly less optimal cluster separation.  

### **Cluster Distribution Analysis**  
- **KMeans Clusters**:  
  - Cluster **0** contains **82** families, making it the dominant group.  
  - Clusters **1, 2, and 3** contain **2, 1, and 6** families, respectively, indicating smaller specialized groups.  
  - The cluster distribution suggests that most families share similar patterns, while a few show unique trends.  

- **Hierarchical Clusters**:  
  - Cluster **0** again dominates with **82** families.  
  - The remaining families are divided into Clusters **1, 2, and 3** (**4, 1, and 4 families** respectively).  
  - This clustering closely follows KMeans, but the silhouette score indicates it is slightly less optimal in maintaining well-separated groups.  

- **DBSCAN Clusters**:  
  - Cluster **0** contains **72** families, indicating strong core group formation,followed byy cluster **1** contains **5families**  
  - **14 families were labeled as noise (-1)**, meaning they did not fit well into any cluster.  
  - The presence of noise suggests that some families exhibit significantly different patterns compared to the main group, which DBSCAN treats as outliers.  

### **Conclusion**  
The clustering results indicate that **Kmean produced the best-defined clusters** with a high silhouette score. **hierarchical  performed well**, creating similar groups but with less silhouette coefficient. **DBSACN Clustering performed the weakest** based on the silhouette score.  

If handling outliers is important, **DBSCAN is the best choice**. If a **strict clustering structure without noise** is needed, **KMeans is a strong alternative**. Hierarchical clustering could still be useful for visualizing relationships but is less optimal for this dataset.

```python
# Select features for clustering
selected_columns = ['FamilleArticle', 'Season']
selected_data = clustering_data[selected_columns]

# Count occurrences of each FamilleArticle per Season
row_counts = selected_data.groupby(['FamilleArticle', 'Season']).size().reset_index(name='RowCounts')

# Create pivot table
pivot_data = row_counts.pivot_table(index='FamilleArticle', columns='Season', values='RowCounts', aggfunc='mean', fill_value=0)

# Normalize the data
scaler = StandardScaler()
normalized_data = scaler.fit_transform(pivot_data)

# Define the number of clusters for KMeans and Hierarchical Clustering
num_clusters = 4  

# Apply KMeans Clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans_clusters = kmeans.fit_predict(normalized_data)
pivot_data['KMeans_Cluster'] = kmeans_clusters

# Apply Hierarchical Clustering
hierarchical_model = AgglomerativeClustering(n_clusters=num_clusters, metric='euclidean', linkage='ward')
hierarchical_clusters = hierarchical_model.fit_predict(normalized_data)
pivot_data['Hierarchical_Cluster'] = hierarchical_clusters

# Apply DBSCAN Clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)  
dbscan_clusters = dbscan.fit_predict(normalized_data)
pivot_data['DBSCAN_Cluster'] = dbscan_clusters

# Function to display the number of unique FamilleArticle in each cluster
def display_cluster_info(cluster_column_name):
    print(f"\n{cluster_column_name} Clusters:")
    for cluster_id, cluster_group in pivot_data.groupby(cluster_column_name):
        num_unique_families = cluster_group.index.nunique()
        print(f"Number of Unique FamilleArticle in Cluster {cluster_id}: {num_unique_families}")

# Display cluster information for each method
display_cluster_info('KMeans_Cluster')
display_cluster_info('Hierarchical_Cluster')
display_cluster_info('DBSCAN_Cluster')
```
