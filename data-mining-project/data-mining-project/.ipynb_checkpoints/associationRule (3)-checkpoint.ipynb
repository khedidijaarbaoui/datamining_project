# Association Rule

## Goal of the Study
The goal of this study is to uncover the relationships between items, specifically identifying which items are frequently associated with one another. By increasing the sales of an antecedent, the sales of the consequent are likely to rise as well, creating a positive feedback loop that drives overall sales growth. We will analyze association rules across different seasons, identifying which patterns consistently appear over the years. This consistency will help us confirm that these associations are not random, but rather meaningful insights that can inform strategic decisions. Moreover, we will explore association rules with varying levels of support and confidence to obtain a more comprehensive understanding of these relationships. We will focus on the autumn and summer seasons due to their lower sales, so by understanding the association rules, we can optimize the sales strategy and boost sales efficiently.s
.


This function is used to divide the dataset by season of the year.

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
from IPython.display import display

# Function to assign seasons based on the month
def get_season(month):
    if month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Autumn'
    else:
        return 'Winter'

# Function to filter the dataset for a specific year and season
def get_seasonal_dataset(df, year, season):
    # Convert 'Date' column to datetime if not already
    if not pd.api.types.is_datetime64_any_dtype(df['Date']):
        df['Date'] = pd.to_datetime(df['Date'])

    # Add year, month, and season columns if not already added
    if 'Year' not in df.columns:
        df['Year'] = df['Date'].dt.year
    if 'Month' not in df.columns:
        df['Month'] = df['Date'].dt.month
    if 'Season' not in df.columns:
        df['Season'] = df['Month'].apply(get_season)

    # Filter dataset by year and season
    filtered_df = df[(df['Year'] == year) & (df['Season'] == season) & (df['TypeEffet'] == 'Bon de livraison')]
    
    # Display the filtered dataset nicely in Jupyter
    #print(f"Filtered Data for year {year} and season {season} (Records: {len(filtered_df)}):")
    #display(filtered_df.head(10))  # Display the first 10 rows for clarity

    return filtered_df

```

the function that generate the association rules

```python
# Function to generate association rules
def generate_association_rules(dataset, min_support=0.7, min_confidence=0.5):
    # Group transactions by 'CleEffet' and aggregate 'FamilleArticle' into a list
    transactions = dataset.groupby('Date')['FamilleArticle'].agg(list).reset_index()

    # Convert the transactions into a list of lists
    transaction_list = transactions['FamilleArticle'].tolist()

    # Use TransactionEncoder to one-hot encode the transactions
    te = TransactionEncoder()
    transaction_matrix = te.fit(transaction_list).transform(transaction_list)
    transaction_matrix_df = pd.DataFrame(transaction_matrix, columns=te.columns_)

    # Apply the apriori algorithm to find frequent itemsets
    frequent_itemsets = apriori(transaction_matrix_df, min_support=min_support, use_colnames=True)

    # Check if there are frequent itemsets
    if frequent_itemsets.empty:
        print("No frequent itemsets found. Try lowering the support threshold.")
        return None, None
    else:
    # Generate association rules
        rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=min_confidence, num_itemsets=len(frequent_itemsets))
    
        return frequent_itemsets, rules


```

Generate the association rules for Winter 2022 and display only the interesting rules that have average support but high confidence.

```python
# Load the dataset
df = pd.read_csv('./Data/Prepocessing2_data.csv')

# Example usage: Filter dataset for Winter 2022 and generate association rules
year = 2022
season = 'Winter'

# Get the seasonal dataset
seasonal_df = get_seasonal_dataset(df, year, season)

# Generate association rules for the seasonal dataset
frequent_itemsets, rules = generate_association_rules(seasonal_df, min_support=0.6, min_confidence=0.8)
print('Frequent itemset: ')
display(frequent_itemsets)

# Filter rules with support less than 0.7 and confidence greater than 0.8
filtered_rules = rules[(rules['support'] < 0.7) & (rules['confidence'] > 0.8)]

# Display the filtered rules
print("Filtered Association Rules (Support < 0.7 and Confidence > 0.8):")
display(filtered_rules)
```

This function determines the association rules that are common in the last 'x' years (where 'x' is a variable that we can input to determine how many years we want to study). It returns only the association rules whose support is between the minimum and maximum support values, and whose confidence is between the minimum and maximum confidence values.

```python
import warnings
warnings.filterwarnings("ignore", category=RuntimeWarning)
def find_common_association_rules_last_X_years(df, season, support_min=0.6, support_max=1, confidence_min=0.8, confidence_max=1, x=0):
    if not pd.api.types.is_datetime64_any_dtype(df['Date']):
        df['Date'] = pd.to_datetime(df['Date'])
    df['Year'] = df['Date'].dt.year
    df['Season'] = df['Date'].dt.month.apply(get_season)

    # Get the last 'x' years in the dataset
    last_X_years = sorted(df['Year'].unique())[-x:]
    print(f"Analyzing common association rules for season: {season}")
    print(f"Last {x} years in dataset: {last_X_years}")

    # Initialize a list to hold association rules for each year
    association_rules_by_year = []

    for year in last_X_years:
        seasonal_df = get_seasonal_dataset(df, year, season)
        if not seasonal_df.empty:
            frequent_itemsets, rules = generate_association_rules(seasonal_df, min_support=support_min, min_confidence=confidence_min)

            
            # Add only the antecedents and consequents to the set for intersection
            rule_tuples = set(zip(rules['antecedents'], rules['consequents']))
            association_rules_by_year.append(rule_tuples)

    # Find the intersection of rules across the last 'x' years
    if association_rules_by_year:
        common_rules = set.intersection(*association_rules_by_year)

        # Convert to DataFrame and filter based on support and confidence range
        full_rules_list = []
        for year in last_X_years:
            seasonal_df = get_seasonal_dataset(df, year, season)
            if not seasonal_df.empty:
                _, rules = generate_association_rules(seasonal_df, min_support=support_min, min_confidence=confidence_min)
              
                full_rules_list.append(rules)

        # Concatenate all years' rules into one DataFrame
        all_rules_df = pd.concat(full_rules_list, ignore_index=True)

        # Filter to keep only common rules within the specified support & confidence range
        filtered_common_rules_df = all_rules_df[
            (all_rules_df.apply(lambda row: (row['antecedents'], row['consequents']) in common_rules, axis=1)) &
            (all_rules_df['support'] >= support_min) & (all_rules_df['support'] <= support_max) &
            (all_rules_df['confidence'] >= confidence_min) & (all_rules_df['confidence'] <= confidence_max)
        ]

        # Remove duplicates for antecedents and consequents (ensuring no repetition)
        filtered_common_rules_df = filtered_common_rules_df.drop_duplicates(subset=['antecedents', 'consequents'])

        if not filtered_common_rules_df.empty:
            print(f"Filtered Common Association Rules (Support: {support_min} - {support_max}, Confidence: {confidence_min} - {confidence_max}):")
            display(filtered_common_rules_df[['antecedents', 'consequents','support','confidence','lift']])
            return filtered_common_rules_df
        else:
            print(f"No common rules met the support ({support_min}-{support_max}) and confidence ({confidence_min}-{confidence_max}) criteria.")
            return pd.DataFrame()
    else:
        print(f"No common association rules found across the last {x} years.")
        return pd.DataFrame()



```

Generating common association rules for Autumn season in the last ten years.

```python
# Example usage:
season = 'Autumn'
common_rules_Autumn = find_common_association_rules_last_X_years(df, season, support_min=0.6, support_max=0.7, confidence_min=0.8, confidence_max=1, x=10)
print(f"the number of rules : {len(common_rules_Autumn)}")
```

```python
common_rules_autumn_list=common_rules_Autumn[['antecedents', 'consequents','support','confidence','lift']].copy()

common_rules_autumn_list['antecedents'] = common_rules_autumn_list['antecedents'].apply(lambda x: ', '.join(list(x)))
common_rules_autumn_list['consequents'] = common_rules_autumn_list['consequents'].apply(lambda x: ', '.join(list(x)))

csv_file_path = './Data/List_Associated_items_autumnSeason.csv'

# Use the to_csv method to write the DataFrame to a CSV file
common_rules_autumn_list.to_csv(csv_file_path, index=False)

print(f'DataFrame has been successfully written to {csv_file_path}')
```

## Plot the Sales Trend Comparison between Antecedent and Consequent Items in the Rules for the Last 5 Years (Autumn and Summer Seasons)

As observed in the clustering analysis, the Autumn and Summer seasons experience lower sales. To address this, we will focus on these seasons to evaluate whether our association rules are effective. By comparing the sales trends of antecedent items and their corresponding consequent items, we aim to assess the relationship between their sales over time.

If the sales trends for both the antecedent and consequent items follow similar patterns, this suggests that the items are indeed correlated in terms of sales performance. In turn, this validates the effectiveness of the association rules, as it shows that boosting sales of antecedent items could potentially lead to an increase in sales for the consequent items, especially during this weak seasons.

```python
import matplotlib.pyplot as plt


def plot_sales_correlation_per_rule_last_X_years(df, common_rules, window=3,x=0):
    """
    Plots separate line graphs for each antecedent-consequent pair
    to show the correlation between their sales trends over the last 5 years,
    with smoothing applied using a rolling mean.
    """
    if common_rules is None or common_rules.empty:
        print("No common association rules found. Exiting plot function.")
        return

    # Get the most recent year from the data and filter the last 5 years
    last_year = df['Year'].max()
    start_year = last_year - x
    df_filtered = df[df['Year'] >= start_year]

    # Convert 'Date' column to datetime and create a 'Month-Year' column
    df_filtered.loc[:, 'Date'] = pd.to_datetime(df_filtered['Date'])
    df_filtered.loc[:, 'Month-Year'] = df_filtered['Date'].dt.to_period('M')

    # Ensure 'Month-Year' is sorted in ascending order
    df_filtered = df_filtered.sort_values(by='Month-Year')

    for index, rule in common_rules.iterrows():
        antecedents = list(rule['antecedents'])
        consequents = list(rule['consequents'])

        # Filter sales for antecedents and consequents
        antecedent_sales = (
            df_filtered[df_filtered['FamilleArticle'].isin(antecedents)]
            .groupby('Month-Year')
            .size()
        )
        consequent_sales = (
            df_filtered[df_filtered['FamilleArticle'].isin(consequents)]
            .groupby('Month-Year')
            .size()
        )

        # Normalize sales for comparison
        antecedent_sales = antecedent_sales / antecedent_sales.max()
        consequent_sales = consequent_sales / consequent_sales.max()

        # Apply rolling mean for smoothing
        smoothed_antecedent_sales = antecedent_sales.rolling(window=window).mean()
        smoothed_consequent_sales = consequent_sales.rolling(window=window).mean()

        # Create the plot
        plt.figure(figsize=(10, 6))
        plt.plot(
            smoothed_antecedent_sales.index.astype(str),
            smoothed_antecedent_sales.values,
            label=f"Antecedent: {', '.join(antecedents)}",
            linestyle="--",
            marker="o",
        )
        plt.plot(
            smoothed_consequent_sales.index.astype(str),
            smoothed_consequent_sales.values,
            label=f"Consequent: {', '.join(consequents)}",
            linestyle="-",
            marker="x",
        )

        # Customize the plot
        plt.title(f"Smoothed Sales Trends (Last 5 Years): {', '.join(antecedents)} → {', '.join(consequents)}", fontsize=14)
        plt.xlabel("Month-Year", fontsize=12)
        plt.ylabel("Normalized Sales (Smoothed)", fontsize=12)
        plt.xticks(rotation=45, fontsize=10)
        plt.legend(loc="best", fontsize=10)
        plt.grid(alpha=0.4)
        plt.tight_layout()

        # Show the plot
        plt.show()
```

```python
# Example: Correlation for Autumn sales trends
plot_sales_correlation_per_rule_last_X_years(df, common_rules_Autumn.head(10), window=4,x=5)

```

```python
# Example: Correlation for Summer sales trends
plot_sales_correlation_per_rule_last_X_years(df, common_rules_Summer.head(10), window=4,x=5)
```
